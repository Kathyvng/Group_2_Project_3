```python
import json
import pandas as pd
```

```python
# Load the JSON file
with open('disease.json') as f:
    data = json.load(f)

# Check the number of data points
num_data_points = len(data)
print(f'The number of data points in the JSON file is: {num_data_points}')
```

```python
disease_df= pd.read_json('disease.json')
disease_df.head(5)
```

```python
state_counts= disease_df['LocationDesc'].value_counts()
state_counts
```

```python
disease_counts = disease_df.groupby(['LocationDesc', 'CDI_Topic']).size().reset_index(name='Count')
disease_counts
```

```python
al_counts = disease_counts.loc[disease_counts['LocationDesc'] == 'Alabama']
al_counts
```

```python
# Calculate the total counts for Alabama
total_counts = al_counts['Count'].sum()

# Calculate the percentage for each disease using .loc
al_counts.loc[:, 'Percentage'] = (al_counts['Count'] / total_counts) * 100

# Round the Percentage to 2 decimal points
al_counts.loc[:,'Percentage'] = al_counts['Percentage'].round(2)

# Display the updated DataFrame with percentages
print(al_counts)
```

```python
al_diseases = al_counts.sort_values(by='Count', ascending=False).head(3)
al_diseases
```

```python
disease_counts = disease_df.groupby(['LocationDesc', 'CDI_Topic']).size().reset_index(name='Count')

# Export to JSON file
disease_counts.to_json('disease_counts.json', orient='records', lines=True)
```
